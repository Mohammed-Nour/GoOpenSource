```python
import urllib.request
import ssl

def download_md_and_convert_to_list(urls):
    readme_contents = []  # Initialize an empty list to hold the content of each README

    for url in urls:
        # Convert GitHub URL to the corresponding raw content URL
        raw_url = url.replace("https://github.com/", "https://raw.githubusercontent.com/").replace("/blob/", "/")

        # Attempt to download the README.md content
        try:
            context = ssl._create_unverified_context()  # Prevent SSL problems
            request = urllib.request.urlopen(raw_url, context=context)
            readme_content = request.read().decode()

            readme_contents.append(readme_content)  # Append the content to the list
            # print(f'Downloaded and added content from URL: {url}')

        except Exception as e:
            readme_contents.append('NaN')
            # print(f'Error downloading content from {url}: {e}')

    return readme_contents

# Example usage:
urls = [
    "https://github.com/fugue/credstash/blob/master/README.md",
    "https://github.com/gliese581gg/YOLO_tensorflow/blob/master/README.md",
    # Add more URLs as needed
]
urls = df['Readme URL modified']
readme_list = download_md_and_convert_to_list(urls)
# readme_list
```

```python
import markdown2
from bs4 import BeautifulSoup

def extract_text_from_md(md_file_path):
    # Read the Markdown file
    with open(md_file_path, 'r', encoding='utf-8') as file:
        md_content = file.read()

    # Convert Markdown to HTML
    html_content = markdown2.markdown(md_content)

    # Parse HTML and extract text
    soup = BeautifulSoup(html_content, 'html.parser')
    text = soup.get_text(separator='\n')

    return text

# Path to your Markdown file
md_file_path = 'your_markdown_file.md'

# Extract text from Markdown file
extracted_text = extract_text_from_md("/content/readme0.txt")

# Print the extracted text
print(extracted_text)
```

```python
def clean_md(md):
  html_content = markdown2.markdown(md)

  # Parse HTML and extract text
  soup = BeautifulSoup(html_content, 'html.parser')
  text = soup.get_text(separator='\n')
  return text
```

```python
def args_sort(*args):
    # Create a list of tuples containing the original index and value
    indexed_args = [(index, value) for index, value in enumerate(args[0])]
    # Sort the list of tuples based on the values
    sorted_args = sorted(indexed_args, key=lambda x: x[1])
    # Extract the original indices from the sorted list
    sorted_indices = [index for index, _ in sorted_args]
    return sorted_indices
```

```python
def find_closest_embeddings(embeddings, chosen_embedding, k=3):
  """
  This function finds the top k closest embeddings to a chosen embedding in a list.

  Args:
      embeddings: A list of numpy arrays representing the embeddings.
      chosen_embedding: The numpy array representing the chosen embedding.
      k: The number of closest embeddings to return (default: 3).

  Returns:
      A list of tuples containing the top k closest embeddings and their distances.
  """
  distances = []  # Store distances between chosen embedding and others
  for embedding in embeddings:
    if embedding is not chosen_embedding:  # Avoid comparing to itself
      distance = np.linalg.norm(chosen_embedding - embedding)  # Calculate distance
      distances.append(distance)
  # Sort by distance (ascending order) and select top k closest
  # sorted_distances = sorted(distances, key=lambda x: x[1])[:k]
  indices = args_sort(distances)
  return indices
```
